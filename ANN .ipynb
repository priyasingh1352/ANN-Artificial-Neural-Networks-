```python
input_data = dataset.iloc[:,:-1]
output_data = dataset.iloc[:,-1:]
```

```python
ss = StandardScaler()
```

```python
input_data = dataset.drop(["RowNumber","CustomerId","Surname","Exited"],axis = 1)

input_data = pd.get_dummies(input_data,drop_first=True)

input_data = pd.DataFrame(ss.fit_transform(input_data),columns = input_data.columns)
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
input_data.head()
```

```python
input_data.shape
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units=6,input_dim = 11, activation = "relu"))
ann.add(Dense(units=4,activation = "relu"))
ann.add(Dense(units=2,activation = "relu"))
ann.add(Dense(units=1,activation = "sigmoid"))
```

```python
ann.compile(optimizer = "adam", loss = "binary_crossentropy" , metrics = ["accuracy"])
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,test_size = 0.2 , random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
print("x_train shape:",x_train.shape)
print("y_train shape:",y_train.shape)
```

```python
print(x_train.shape,y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
ann.add(Dense(1,activation = "sigmoid"))
ann.compile(optimizer = "adam", loss = "binary_crossentropy" , metrics = ["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=80,epochs = 100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ann.predict(x_test)
```

```python
prd = ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
    if i[0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
prd1 = ann.predict(x_train)
prd_data1 = []
for i in prd1:
    if i[0]>0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
accuracy_score(y_train,prd_data1)*100
```
